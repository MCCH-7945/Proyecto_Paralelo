````python
# Opcional: vista rápida con pocas muestras (no genera los CSV grandes)
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

n_demo = 5000
pts, _ = make_blobs(
    n_samples=n_demo,
    centers=8,
    n_features=2,
    cluster_std=0.04,
    random_state=7,
    center_box=(0.0, 1.0),
)
pts = np.round(np.abs(pts), 3)
plt.figure(figsize=(7, 6))
plt.scatter(pts[:, 0], pts[:, 1], s=3, alpha=0.35)
plt.title("Vista 2D de muestra (demo, no es un archivo de salida)")
plt.xlabel("x")
plt.ylabel("y")
plt.axis("equal")
plt.grid(True, alpha=0.3)
plt.show()
````
